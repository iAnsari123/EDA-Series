1) Instructions (For Data Cleaning)
> __Find all Null values in the dataset. If there is any null values in any column, then fill it with the mean of that column.__

```python
df.isna().apply(pd.value_counts)
```

```python
df = df.dropna(how='all')
```

```python
df['Cylinders'] = df['Cylinders'].fillna(df['Cylinders'].mean())
```

2. Questions (Based on value counts)
> __Check what are the different types of make are there in our dataset. And, what is the count(occurences) of each make in the data?__

```python
df['Make'].nunique()
```

```python
df['Make'].value_counts()
```

3. Instructions (Filtering)
> __Show all the records where origin is Aisa or Europe__

```python
df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')]
```

4. Instructions (Removing unwanted records)
> __Remove all the records (rows) where weight is above 4000__

```python
df[~(df['Weight'] > 4000)]
```

5. Instruction (applying function on a column)
> __Increase all the values of 'MPG_City' column by 3.__

```python
df['MPG_City'] = df['MPG_City'].apply(lambda x: x + 3)
```
