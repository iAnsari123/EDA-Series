### Q. 4) Calculate state-wise :

### A. Total number of population.

```python
f = df.groupby(['State_name'])[['Population']].sum()
f.sort_values(by='Population',ascending=False)
```

### B. Total no. of the population with different religions.

```python
df.groupby(['State_name'])[['Muslims','Hindus','Christians','Sikhs','Buddhists','Jains']].sum()
```

### Q. 5) How many Male Workers were there in Maharashtra state ?

```python
df[df['State_name'] == "MAHARASHTRA"][['Male_Workers']].sum()
df[df['State_name'] == "MAHARASHTRA"][['Female_Workers']].sum()
```
